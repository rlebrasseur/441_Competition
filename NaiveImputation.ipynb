```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
df = df.drop(['uniqueid'], axis=1)
df.head()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

# naive data processing using mean for imputation, and not differentiating between numerical and categorical data.

X = df.drop(["health"], axis=1)
X = X.values
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(X)
X = imp.transform(X)
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
y = df['health']
y = y.values
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

# neural network trained on this naively processed data.

size_hl = int(np.floor((len(df.columns)-1+5)/2))
model = MLPClassifier(random_state=20803652, hidden_layer_sizes=(size_hl), max_iter=1000)
param_grid = dict(learning_rate_init=[0.001, 0.01, 0.1, 0.2], alpha=[0, 0.0001, 0.001, 0.01])
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring="neg_log_loss", n_jobs=-1, verbose=1)
grid_result = grid.fit(X, y)
print(grid_result.best_score_)
print(grid_result.best_params_)
```
